```python
"""
given a list of numbers i.e. [1,2,4,-5,7,9,3,2], make another list that contains all the 
items in sorted order from min to max. i.e your results will be another list [-5,1,2,2,3,3,7,9]
"""

L = [1,2,4,-5,7,9,3,2]
m = L[0]
idx = 0 
c=0
for i in L: 
    if i<m:
        m=i
        idx=c
    c+=1
tmp = L[0]
L[0] = m 
L[idx] = tmp
print(L)
```

```python
L = [1,2,4,-5,7,9,3,2]
for j in range(len(L)):
    m = L[j]
    idx = j 
    c = j 
    for i in range(j,len(L)):
        if L[i]<m:
            m = L[i]
            idx = c
        c+=1
    tmp = L[j]
    L[j] = m
    L[idx] = tmp
print(L)
```

```python
def findMin(L,startIndx):
    m = L[startIndx]
    idx = startIndx
    for i in range(startIndx,len(L)):
        x = L[i]
        if i<m:
            m = x
            idx = i
        else:
            pass
    return m,idx
```

```python
a,b = findMin([2,3,4,0,9])
```

```python
def swapvalues(L,idx1,idx2):
    tmp = L[idx1]
    L[idx1] = L[idx2]
    L[idx2] = tmp
    return L
```

```python
L=[2,3,6,4]
L2 = swapvalues(L,2,3)
print(L2)
```
