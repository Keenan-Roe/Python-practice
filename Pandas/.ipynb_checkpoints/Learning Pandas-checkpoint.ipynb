```python
grades_dict = {'A':4,'B':3.5,'C':3,'D':2.5}
grades = pd.Series(grads_dict)
```

```python
marks
```

```python
marks['A']
```

```python
marks[0:2]
```

```python
D = pd.DataFrame({'marks':marks,'grades':grades})
```
